```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import datetime
```

```python
train=pd.read_csv("Train.csv")
test=pd.read_csv("Test.csv")
```

```python
dummy=pd.get_dummies(train[['weather_type']])
```

```python
train=pd.concat([train,dummy],axis=1)
```

```python
train.head()
```

```python
train.drop(["weather_type"],axis=1,inplace=True)
```

```python
for index,row in train.iterrows():
    if row["weather_type_Squall"]==1:
        train.at[index,"weather_type_Rain"]=1
        train.at[index,"weather_type_Snow"]=1
        train.at[index,"weather_type_Thunderstorm"]=1
    if row["weather_type_Smoke"]==1:
        train.at[index,"weather_type_Haze"]=1
```

```python
for index,row in train.iterrows():
    ##Clear Weather Mapping
    if(row["weather_description"]=="sky is clear"):
        train.at[index,"weather_type_Clear"]=1
    if(row["weather_description"]=="Sky is Clear"):
        train.at[index,"weather_type_Clear"]=2
    ##Clouds Mapping
    if(row["weather_description"]=="few clouds"):
        train.at[index,"weather_type_Clouds"]=3
    if(row["weather_description"]=="scattered clouds"):
        train.at[index,"weather_type_Clouds"]=4
    if(row["weather_description"]=="broken clouds"):
        train.at[index,"weather_type_Clouds"]=2
    if(row["weather_description"]=="overcast clouds"):
        train.at[index,"weather_type_Clouds"]=1
    ##Drizzle Mapping
    if(row["weather_description"]=="light intensity drizzle"):
        train.at[index,"weather_type_Drizzle"]=4
    if(row["weather_description"]=="drizzle"):
        train.at[index,"weather_type_Drizzle"]=2
    if(row["weather_description"]=="heavy intensity drizzle"):
        train.at[index,"weather_type_Drizzle"]=3
    if(row["weather_description"]=="shower drizzle"):
        train.at[index,"weather_type_Drizzle"]=1
    ##rain Mapping
    if(row["weather_description"]=="light rain"):
        train.at[index,"weather_type_Rain"]=4
    if(row["weather_description"]=="moderate rain"):
        train.at[index,"weather_type_Rain"]=3
    if(row["weather_description"]=="proximity shower rain"):
        train.at[index,"weather_type_Rain"]=7
    if(row["weather_description"]=="heavy intensity rain"):
        train.at[index,"weather_type_Rain"]=8
    if(row["weather_description"]=="light intensity shower rain"):
        train.at[index,"weather_type_Rain"]=5
    if(row["weather_description"]=="very heavy rain"):
        train.at[index,"weather_type_Rain"]=1
    if(row["weather_description"]=="freezing rain"):
        train.at[index,"weather_type_Rain"]=6
    if(row["weather_description"]=="sleet"):
        train.at[index,"weather_type_Rain"]=5
    ##snow mapping
    if(row["weather_description"]=="SQUALLS"):
        train.at[index,"weather_type_Snow"]=1
    if(row["weather_description"]=="snow"):
        train.at[index,"weather_type_Snow"]=2
    if(row["weather_description"]=="light rain and snow"):
        train.at[index,"weather_type_Snow"]=5
    if(row["weather_description"]=="light snow"):
        train.at[index,"weather_type_Snow"]=3
    if(row["weather_description"]=="light shower snow"):
        train.at[index,"weather_type_Snow"]=6    
    if(row["weather_description"]=="shower snow"):
        train.at[index,"weather_type_Snow"]=7
    if(row["weather_description"]=="heavy snow"):
        train.at[index,"weather_type_Snow"]=4
    ##Thunderstorm mapping
    if(row["weather_description"]=="thunderstorm"):
        train.at[index,"weather_type_Thunderstorm"]=5
    if(row["weather_description"]=="proximity thunderstorm"):
        train.at[index,"weather_type_Thunderstorm"]=8
    if(row["weather_description"]=="proximity thunderstorm with rain"):
        train.at[index,"weather_type_Thunderstorm"]=1
    if(row["weather_description"]=="proximity thunderstorm with drizzle"):
        train.at[index,"weather_type_Thunderstorm"]=6
    if(row["weather_description"]=="thunderstorm with light drizzle"):
        train.at[index,"weather_type_Thunderstorm"]=7    
    if(row["weather_description"]=="thunderstorm with drizzle"):
        train.at[index,"weather_type_Thunderstorm"]=2
    if(row["weather_description"]=="thunderstorm with light rain"):
        train.at[index,"weather_type_Thunderstorm"]=3
    if(row["weather_description"]=="thunderstorm with rain"):
        train.at[index,"weather_type_Thunderstorm"]=9
    if(row["weather_description"]=="thunderstorm with heavy rain"):
        train.at[index,"weather_type_Thunderstorm"]=4
    ##Fog Mapping
    if(row["weather_description"]=="fog"):
        train.at[index,"weather_type_Fog"]=1
    ##Mist Mapping
    if(row["weather_description"]=="mist"):
        train.at[index,"weather_type_Mist"]=1
    ##Smoke Mapping
    if(row["weather_description"]=="smoke"):
        train.at[index,"weather_type_Haze"]=1
    if(row["weather_description"]=="haze"):
        train.at[index,"weather_type_Haze"]=2
```

```python
train.drop(["weather_type_Squall","weather_type_Smoke"],inplace=True,axis=1)
```

```python
for index,row in train.iterrows():
    train.at[index,"date"]=datetime.datetime.strptime(row["date_time"], "%Y-%m-%d %H:%M:%S").date()
    train.at[index,"time"]=datetime.datetime.strptime(row["date_time"], "%Y-%m-%d %H:%M:%S").time()
    train.at[index,"hour"]=datetime.datetime.strptime(row["date_time"], "%Y-%m-%d %H:%M:%S").time().hour
    train.at[index,"month"]=datetime.datetime.strptime(row["date_time"], "%Y-%m-%d %H:%M:%S").date().month
    train.at[index,"day"]=datetime.datetime.strptime(row["date_time"], "%Y-%m-%d %H:%M:%S").date().day
    train.at[index,"year"]=datetime.datetime.strptime(row["date_time"], "%Y-%m-%d %H:%M:%S").date().year
```

```python
train['date_time'] = pd.to_datetime(train.date_time)
train['dayofyear'] = train['date_time'].dt.dayofyear
train['weekofyear'] = train['date_time'].dt.weekofyear
train['dayofweek'] = train['date_time'].dt.dayofweek
```

```python
train.drop(["date_time","weather_description"],axis=1,inplace=True)
```

```python
train.head()
```

```python
train.shape
```

```python
train["is_holiday"].unique()
```

```python
vard = 0
varm = 0
var = ""
for index,row in train.iterrows():
    if(row['is_holiday'] != "None"):
        vard = row['day']
        varm = row['month']
        var = row['is_holiday']
    if(row['day'] == vard and row['month'] == varm):
        train.loc[index,'is_holiday']=var
```

```python
for index,row in train.iterrows():
    if row["is_holiday"]!="None":
        train.at[index,"is_holiday"]=1
    else:
        train.at[index,"is_holiday"]=0
```

```python
train["is_holiday"]=train["is_holiday"].astype(int)
```

```python
train.drop(["air_pollution_index","wind_speed","wind_direction","visibility_in_miles","dew_point","rain_p_h","snow_p_h","year","humidity"],axis=1,inplace=True)
```

```python
train.head()
```

```python
train.drop(["clouds_all"],inplace=True,axis=1)
```

```python
train.head()
```

```python
for index,row in train.iterrows():
    if(row['hour'] >9 and row['hour'] <16):
        train.loc[index,'peak_time'] = 1
    else:
        train.loc[index,'peak_time'] = 4
```

```python
y=train["traffic_volume"]
```

```python
train.drop(["traffic_volume"],axis=1,inplace=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(train,y,test_size=0.1,shuffle=True)
```

```python
model=RandomForestRegressor(bootstrap=True,n_estimators=100)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
# x_train.drop(["date","time"],inplace=True,axis=1)
# x_test.drop(["date","time"],inplace=True,axis=1)
x_train.drop(["time"],inplace=True,axis=1)
x_test.drop(["time"],inplace=True,axis=1)
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_log_error
```

```python
mean_absolute_error(y_test,prediction)
```

```python
np.sqrt(mean_squared_log_error(y_test,prediction))
```
